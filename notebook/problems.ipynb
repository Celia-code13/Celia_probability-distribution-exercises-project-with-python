## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# 1) Definimos todas las funciones de probabilidad relacionadas con la DISTRIBUCIÓN BINOMIAL

def dbinom(x, size, prob = 0.5): # Recordamos que aquí este 0.5 es un valor predeterminado, es decir, el valor que aparecerá si no especificamos la probabilidad.
    
    # Con BINOM.PMF vamos a calcular la función de masa de probabilidad (PMF) de una distribución binomial para un valor dado "x", donde "size" es el tamaño de la muestra y "prob" es la probabilidad de éxito de en cada ensayo.
    # El parámetro "loc" no es necesario en una distribución binomial

    result = binom.pmf(k = x, n = size, p = prob, loc = 0)  # Recordar que la x no tiene que ser necesariamente un valor puntual, puede ser un rango de datos, muy útil para plotear.
    return result

def pbinom(q, size, prob = 0.5):

    # Con BINOM.CDF calculamos la función de distribución acumulativa (CDF) de la distribución binomial hasta un valor específico k
    # Representa la acumulación de probabilidades desde el inicio de la distribución hasta el valor k.
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):

    # Con BINOM.PPF calculamos la función cuantil de la distribución binomial.
    # Se supone que la ppf es la función inversa de la CDF. Mientras que la CDF nos da la probabilidad acumulada hasta un valor k/x, la PPF nos da el valor correspondiente para una cierta probabilidad acum.
 
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):

    # Con BINOM.RVS podemos generar variables aleatorias en la distribución binomial. 
    # En este caso, k (primer argumento) es el número de variables que se desea generar.
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# 2) Aplicamos las distintas funciones con una valores dados comunes (n o size = 10, como número de ensayos, y p = 0.2, como la probabilidad de éxito de cada ensayo) para una distribución binomial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python

# 3) Utilizamos MatplotLib para visualizar la función de masa de probabilidad (PMF), la función de distribución acumulativa (CDF) y la función cuantil (PPF).

import matplotlib.pyplot as plt

np.random.seed(42) # Garantiza que no cambien los números aleatorios generados. El número que se escriba es irrelevante.

size = 10 # n
prob = 0.2 # p
data = rbinom(2000, size = 10, prob = 0.2) # k, aunque ahora no es un valor, si no un intervalo de valores generados aleatoriamente gracias a la función binom.rvs

# Definimos las variables para calcular cada función con los datos aportados anteriormente

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

# Creamos una figura con 3 subgráficos dispuestos verticalmente 

fig, axis = plt.subplots(3, 1, figsize = (5, 7))  # 3 filas y 1 columna

# Seleccionamos cada subgráfico de la lista "axis" y utilizamos stem para hacer gráficos de tallos y hojas, donde "x" son los datos y el eje "y" son los resultados de cada una de las funciones

axis[0].stem(data, pmf, "r-") 
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout() # Ajustamos el diseño para que no haya superposiciones

plt.show()
```
